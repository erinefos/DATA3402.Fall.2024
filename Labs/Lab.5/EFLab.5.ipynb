# Quiz 2

Write a function `make_deck` that returns a list of all of the cards in a standard card deck. The return should be a list of tuples of pairs of suit and value. For example the 10 of Clubs would be  `('Clubs', 10)` and Queen of Hearts would be `('Hearts', 'Queen')`. Recall that a deck has 52 cards, divided into 4 suits (Clubs, Diamonds, Hearts, and Spades), and that each suit has 13 cards: 2 to 10, Jack, Queen, King, and Ace. Summit your solution with Lab 5.

```python
def make_deck():
    suits = ['Hearts', 'Spades', 'Diamonds', 'Clubs']
    values = list(range(2,11)) + ['Jack', 'Queen', 'King', 'Ace']
    return [(suit, value) for suit in suits for value in values]
```

1. Create a `matrix` class with the following properties:
    * It can be initialized in 2 ways:
        1. with arguments `n` and `m`, the size of the matrix. A newly instanciated matrix will contain all zeros.
        2. with a list of lists of values. Note that since we are using lists of lists to implement matrices, it is possible that not all rows have the same number of columns. Test explicitly that the matrix is properly specified.
    * Matrix instances `M` can be indexed with `M[i][j]` and `M[i,j]`.
    * Matrix assignment works in 2 ways:
        1. If `M_1` and `M_2` are `matrix` instances `M_1=M_2` sets the values of `M_1` to those of `M_2`, if they are the same size. Error otherwise.
        2. In example above `M_2` can be a list of lists of correct size.

```python
class matrix:
    def __init__(self, n = None, m = None, values = None):
        # Two ways to initialize a matrix
        # First: a list of lists of values
        if values:
            # The lists need to be completely filled in, no missing columns
            # eg. [[0, 1], [0], [0, 1]] would raise the error
            if not all(len(row) == len(values[0]) for row in values):
                raise ValueError('All rows must have the same number of columns.')
            self.values = values
            # Assign n and m based off the given list of lists of values
            self.n = len(values)
            self.m = len(values[0])
        # Second: giving the number of rows and columns and create a matrix of zeros
        else:
            if n is None or m is None:
                raise ValueError('n and m must be specified if values are not given.')
            self.n = n
            self.m = m
            # Assigns the values of the n by m matrix as 0
            self.values = [[0 for _ in range(m)] for _ in range(n)]

    # Support indexing using either M[i][j] or M[i, j]
    def __getitem__(self, index):
        if isinstance(index, tuple):
            row, col = index
            return self.values[row][col]
        return self.values[index]

    def __setitem__(self, index, value):
        if isinstance(index, tuple):
            row, col = index
            self.values[row][col] = value
        else:
            self.values[index] = value

    # Checks to see if two matrices are equal (same size and values)
    def __eq__(self, other):
        if not isinstance(other, matrix):
            return False
        return self.values == other.values

    def __repr__(self):
        return '\n'.join([' '.join(map(str, row)) for row in self.values])

    # Matrix assignment
    def assignment(self, other):
        # Checks if two matrices are the same size
        if isinstance(other, matrix):
            if self.n != other.n or self.m != other.m:
                raise ValueError('Matrices are different size.')
            # Assigns the self matrix the values of the other matrix
            self.values = [row[:] for row in other.values]
        # Checks if a matrix and a list are the same size
        elif isinstance(other, list):
            if not all(len(row) == self.m for row in other):
                raise ValueError('List does not match matrix size.')
            if len(other) != self.n:
                raise ValueError('List does not have the correct number of rows.')
            # Assigns the matrix to the values of the list
            self.values = other
        # Makes sure that no other input but a matrix or a list is accepted
        else:
            raise TypeError('Acceptable inputs are a matrix or a list.')
```

3. Write functions that create special matrices (note these are standalone functions, not member functions of your `matrix` class):
    * `constant(n,m,c)`: returns a `n` by `m` matrix filled with floats of value `c`.
    * `zeros(n,m)` and `ones(n,m)`: return `n` by `m` matrices filled with floats of value `0` and `1`, respectively.
    * `eye(n)`: returns the n by n identity matrix.

```python
# Gives you a n x m matrix filled with the float versions of a specified constant
def constant(n, m, c):
    values = [[float(c) for _ in range(m)] for _ in range(n)]
    return matrix(values = values)

# Using the constant function, gives you a matrix of float zeros
def zeros(n, m):
    return constant(n, m, 0)

# Using the constant function, gives you a matrix of float ones
def ones(n, m):
    return constant(n, m, 1)

# Gives you a matrix with ones in the diagnol and zeros everywhere else
def eye(n):
    values = [[1.0 if i == j else 0.0 for j in range(n)] for i in range(n)]
    return matrix(values = values)
```

```python
class matrix:
    def __init__(self, n = None, m = None, values = None):
        # Two ways to initialize a matrix
        # First: a list of lists of values
        if values:
            # The lists need to be completely filled in, no missing columns
            # eg. [[0, 1], [0], [0, 1]] would raise the error
            if not all(len(row) == len(values[0]) for row in values):
                raise ValueError('All rows must have the same number of columns.')
            self.values = values
            # Assign n and m based off the given list of lists of values
            self.n = len(values)
            self.m = len(values[0])
        # Second: giving the number of rows and columns and create a matrix of zeros
        else:
            if n is None or m is None:
                raise ValueError('n and m must be specified if values are not given.')
            self.n = n
            self.m = m
            # Assigns the values of the n by m matrix as 0
            self.values = [[0 for _ in range(m)] for _ in range(n)]

    # Support indexing using either M[i][j] or M[i, j]
    def __getitem__(self, index):
        if isinstance(index, tuple):
            row, col = index
            return self.values[row][col]
        return self.values[index]

    def __setitem__(self, index, value):
        if isinstance(index, tuple):
            row, col = index
            self.values[row][col] = value
        else:
            self.values[index] = value

    # Checks to see if two matrices are equal (same size and values)
    def __eq__(self, other):
        if not isinstance(other, matrix):
            return False
        return self.values == other.values

    def __repr__(self):
        return '\n'.join([' '.join(map(str, row)) for row in self.values])

    # Matrix assignment
    def assignment(self, other):
        # Checks if two matrices are the same size
        if isinstance(other, matrix):
            if self.n != other.n or self.m != other.m:
                raise ValueError('Matrices are different size.')
            # Assigns the self matrix the values of the other matrix
            self.values = [row[:] for row in other.values]
        # Checks if a matrix and a list are the same size
        elif isinstance(other, list):
            if not all(len(row) == self.m for row in other):
                raise ValueError('List does not match matrix size.')
            if len(other) != self.n:
                raise ValueError('List does not have the correct number of rows.')
            # Assigns the matrix to the values of the list
            self.values = other
        # Makes sure that no other input but a matrix or a list is accepted
        else:
            raise TypeError('Acceptable inputs are a matrix or a list.')

    # Gives you the shape of the matrix (rows, columns)
    def shape(self):
        return (self.n, self.m)

    # Gives you a transposed version of the matrix
    # Eg. the number of rows is now the number of columns and vice versa
    def transpose(self):
        transposed_values = [[self.values[i][j] for j in range(self.n)] for i in range(self.m)]
        return matrix(values = transposed_values)

    # Gives you the specified row you ask for of a matrix as a different matrix
    def row(self, n):
        if n >= self.n or n < 0:
            raise IndexError('Out of index range.')
        return matrix(values=[self.values[n]])

    # Gives you the specified column you ask for of a matrix as a different matrix
    def column(self, n):
        if n >= self.m or n < 0:
            raise IndexError('Out of index range.')
        column_values = [[self.values[i][n] for i in range(self.n)]]
        return matrix(values = column_values)

    # Gives you the matrix as a list
    def to_list(self):
        return self.values

    # Gives you a small version of your matrix
    def block(self, n_0, n_1, m_0, m_1):
        if n_0 < 0 or n_1 > self.n or m_0 < 0 or m_1 > self.m or n_0 > n_1 or m_0 > m_1:
            raise IndexError('Bad block indices')
        block_values = [row[m_0:m_1] for row in self.values[n_0:n_1]]
        return matrix(block_values)

    # Multiplies a matrix by a constant
    def scalarmul(self, c):
        return matrix(values = [[c * self.values[i][j] for j in range(self.m)] for i in range(self.n)])

    # Adds two matrices together
    def add(self, N):
        # Returns an error if the matrices are not the same size
        if self.shape() != N.shape():
            raise ValueError('Matrices are not the same shape.')
        return matrix(values = [[self.values[i][j] + N[i,j] for j in range(self.m)] for i in range(self.n)])

    # Subtracts a matrix from another
    def sub(self, N):
        # Returns an error if the matrices are not the same size
        if self.shape() != N.shape():
            raise ValueError('Matrices are not the same shape.')
        return matrix(values = [[self.values[i][j] - N[i,j] for j in range(self.m)] for i in range(self.n)])

    # Does matrix multiplication
    def mat_mult(self, N):
        if self.m != N.n:
            raise ValueError('Number of columns on one must equal the number of rows on the other.')
        product_values = [[sum(self.values[i][k] * N[k,j] for k in range(self.m)) for j in range(N.n)] for i in range(self.n)]
        return matrix(values = product_values)

    # Does multiplication on the elements
    def element_mult(self, N):
        # Returns an error if the matrices are not the same size
        if self.shape() != N.shape():
            raise ValueError('Matrices are not the same shape')
        return matrix(values = [[self.values[i][j] * N[i, j] for j in range(self.m)] for i in range(self.n)])

    # Checks if two matrices are equal to one another    
    def equals(self, N):
        # Returns an error if the matrices are not the same size
        if self.shape() != N.shape():
            return False
        return all(self.values[i][j] == N[i,j] for i in range(self.n) for j in range(self.m))

    # OVERLOADING FUNCTIONS

    # Addition
    def __add__(self, other):
        # Makes sure the two types being added together are matrices
        if isinstance(other, matrix):
            return self.add(other)
        else:
            raise TypeError('Type must be in matrix form.')

    # Subtraction
    def __sub__(self, other):
        if isinstance(other, matrix):
            return self.sub(other)
        else:
            raise TypeError('Type must be in matrix form.')

    # Mulitplication
    def __mul__(self, other):
        # Scalar mulitplication
        if isinstance(other, (int, float)):
            return self.scalarmul(other)
        # Matrix mulitplication
        elif isinstance(other, matrix):
            return self.mat_mult(other)
        else:
            raise TypeError('Check types.')

    # Right side multiplication because that is a thing
    def __rmul__(self, other):
        # Scalar multiplication
        if isinstance(other, (int, float)):
            return self.scalarmul(other)
        else:
            raise TypeError('Check types.')
```
