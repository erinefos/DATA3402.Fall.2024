1. Write a "counter" class that can be incremented up to a specified maximum value, will print an error if an attempt is made to increment beyond that value, and allows reseting the counter. 

```python
class counter:
    def __init__(self, max_value):
        self.max_value = max_value
        self.value = 0

    # If the value is less than the max_value, it adds 1
    def increment(self):
        if self.value < self.max_value:
            self.value += 1
        # Then prints an error once the value equals the maximum
        else:
            print("Error: Maximum value reached.")

    # Gives you the value of the counter
    def __str__(self):
        return f"Counter value: {self.value}"
        
    # Resets the counter
    def reset(self):
        self.value = 0
```

9. Add a function in the base class of the object classes that returns true/false testing that the object overlaps with another object.

```python
from abc import ABC, abstractmethod
import math

# Implementing a common base class, which subclasses will share functions from
class shape(ABC):
    def area(self):
        pass
        
    def perimeter(self):
        pass

    def coordinates(self):
        pass

    def inside_shape(self, x, y):
        pass

    # New function to find overlapping objects
    def overlaps(self):
        pass

# RECTANGLE
class rectangle(shape):
    def __init__(self, length, width, x, y):
        self.__length = length
        self.__width = width
        self.__x = x
        self.__y = y

    # Computes the area
    def area(self):
        return self.__width * self.__length

    # Computes the perimeter
    def perimeter(self):
        return 2 * (self.__width + self.__length)

    # Computes the coordinates of the shape, starting from the bottom left corner so we are adding length and width
    def coordinates(self):
        bottom_left = (self.__x, self.__y)
        bottom_right = (self.__x + self.__length, self.__y)
        top_right = (self.__x + self.__length, self.__y + self.__width)
        top_left = (self.__x, self.__y + self.__width)
        
        return [bottom_left, bottom_right, top_right, bottom_left]

    # Confirms whether or not a coordinate lies inside a shape or not
    def inside_points(self, x, y):
        # A point must be greater than or equal to (x,y) but less than or equal to (x+length, y+width)
        return (self.__x <= x <= self.__x + self.__length) and (self.__y <= y <= self.__y + self.__width)

    # Confirms whether or not two shapes overlap with one another
    def overlaps(self, other):
        # Checks if the rectangle and another rectangle overlap
        if isinstance(other, rectangle):
            return not (self.__x + self.__length < other.print_x() or
                        self.__x > other.print_x() + other.__length or
                        self.__y + self.__width < other.print_y() or
                        self.__y > other.print_y() + other.__width)
        # Checks if the rectangle and a circle overlap
        elif isinstance(other, circle):
            return any(other.inside_points(px, py) for px, py in self.coordinates()) or \
                   any(self.inside_points(px, py) for px, py in other.coordinates())
        # Checks other shapes using pervious methods
        elif isinstance(other, triangle):
            return any(other.inside_points(px, py) for px, py in self.coordinates()) or \
                   any(self.inside_points(px, py) for px, py in other.coordinates())
        else:
            return False
            
    # Retrieves the values of the data members
    def print_length(self):
        return self.__length

    def print_width(self):
        return self.__width

    def print_x(self):
        return self.__x

    def print_y(self):
        return self.__y

# CIRCLE
class circle(shape):
    def __init__(self, radius, x, y):
        self.__radius = radius
        self.__x = x
        self.__y = y

    # Computes the area
    def area(self):
        return math.pi * (self.__radius ** 2)

    # Computes the perimeter
    def perimeter(self):
        return 2 * math.pi * self.__radius

    # Computes the coordinates of the shape, starting from the center
    def coordinates(self):
        # Using a loop to find 16 points on the circle
        points = []
        num_points = 16
        for i in range(num_points):
            angle = (2 * math.pi / num_points) * i
            # Creating the points by adding the radius, which puts us on the outline and the increments the coordinates by the angle between points
            x = self.__x + self.__radius * math.cos(angle)
            y = self.__y + self.__radius * math.sin(angle)
            points.append((x, y))
        return points
        
    # Confirms whether or not a coordinate lies inside a shape or not
    def inside_points(self, x, y):
        # Computes the distance from the given point from the center point
        distance = math.sqrt((x - self.__x) ** 2 + (y - self.__y) ** 2)
        # The given point must be between the center of the circle and the outside (radius) (less than the radius)
        return distance <= self.__radius

    # Confirms whether or not two shapes overlap with one another
    def overlaps(self, other):
        # Checks if the circle and another circle overlap
        if isinstance(self, circle):
            # Modifying the distance formula to calculate the distance between the centers of the circles
            distance = math.sqrt((self.__x - other.print_x()) ** 2 + (self.__y - other.print_y()) ** 2)
            # If the distance between the centers is less than or equal to the two radii then the circles overlap
            return distance <= self.__radius + other.__radius
        # Checks other shapes using pervious methods
        elif isinstance(other, rectangle):
            closest_x = max(other.print_x(), min(self.__x, other.print_x() + other.print_length()))
            closest_y = max(other.print_y(), min(self.__y, other.print_y() + other.print_width()))
            # Calculate distance from circle center to this closest point
            distance = math.sqrt((closest_x - self.__x) ** 2 + (closest_y - self.__y) ** 2)
            # If the the distance is less than or equal to the radius then they overlap
            return distance <= self.__radius
        elif isinstance(other, triangle):
            return other.overlaps(self)
        else:
            return False
            
    # Retrieves the value of the data members
    def print_radius(self):
        return self.__radius

    def print_x(self):
        return self.__x

    def print_y(self):
        return self.__y

# TRIANGLE
class triangle(shape):
    def __init__(self, base, height, x, y):
        self.__base = base
        self.__height = height
        self.__x = x
        self.__y = y

    # Compute the area
    def area(self):
        return 0.5 * self.__base * self.__height
        
    # Compute the perimeter by using the side_length function
    def perimeter(self):
        side_length = math.sqrt((self.__base / 2) ** 2 + self.__height ** 2)
        return self.__base + 2 * side_length

    # Computes the coordinates of the shape, assuming we start at the left most corner
    def coordinates(self):
        left_vertex = (self.__x, self.__y)
        right_vertex = (self.__x + self.__base, self.__y)
        top_vertex = ((self.__x + self.__base) / 2, self.__y + self.__height)

        return [left_vertex,
                right_vertex,
                top_vertex]

    # Confirms whether or not a coordinate lies inside a shape or not
    def inside_points(self, x, y):
        # Similiar to the coordinates function, however this separates them into their respective placement (x or y)
        # Bottom left x and y values
        x1, y1 = self.__x, self.__y
        # Bottom right x and y values
        x2, y2 = self.__x + self.__base, self.__y
        # Top x and y values
        x3, y3 = self.__x + self.__base / 2, self.__y + self.__height

        # Then we calculate the area of the entire triangle
        area_main = abs((x1*(y2-y3) + x2*(y3-y1) + x3*(y1-y2)) / 2.0)
        # And compute the area of the triangles where we replace a point (1, 2, and 3) with the point we are testing
        area1 = abs((x*(y2-y3) + x2*(y3-y) + x3*(y-y2)) / 2.0)
        area2 = abs((x1*(y-y3) + x*(y3-y1) + x3*(y1-y)) / 2.0)
        area3 = abs((x1*(y2-y) + x2*(y-y1) + x*(y1-y2)) / 2.0)

        # The total area will equal the sub areas if the point is inside the triangle
        if (area_main == area1 + area2 + area3):
            return True
        else:
            return False

    # Confirms whether or not two shapes overlap with one another
    def overlaps(self, other):
        # Checks if the triangle and another triangle overlap
        if isinstance(other, triangle):
            return any(other.inside_points(px, py) for px, py in self.coordinates()) or \
                   any(self.inside_points(px, py) for px, py in other.coordinates())
        # Checks if the triangle and a rectangle overlap
        elif isinstance(other, rectangle):
            return any(self.inside_points(px, py) for px, py in [(other.print_x(), other.print_y()),
                                                                 (other.print_x() + other.print_length(), other.print_y()),
                                                                 (other.print_x(), other.print_y() + other.print_width()),
                                                                 (other.print_x() + other.print_length(), other.print_y() + other.print_width())
                                                                ])
        # Checks if the triangle and any other shape overlap using previous method
        elif isinstance(other, circle):
            return any(other.inside_points(px, py) for px, py in self.coordinates()) or \
                   self.inside_points(other.print_x(), other.print_y())
        else:
            return False
            
    # Retrieve the values of the data members
    def print_base(self):
        return self.__base

    def print_height(self):
        return self.__height

    def print_x(self):
        return self.__x

    def print_y(self):
        return self.__y
```

```python
class canvas:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        # Empty canvas is a matrix with element being the "space" character
        self.data = [[' '] * width for i in range(height)]

    def set_pixel(self, row, col, char='*'):
        self.data[row][col] = char

    def get_pixel(self, row, col):
        return self.data[row][col]
    
    def clear_canvas(self):
        self.data = [[' '] * self.width for i in range(self.height)]
    
    def v_line(self, x, y, height, char = '*'):
        for i in range(height):
            self.set_pixel(x + i, y, char = char)

    def h_line(self, x, y, width, char = '*'):
        for i in range(width):
            self.set_pixel(x, y + i, char = char)
            
    def line(self, x1, y1, x2, y2, char = '*'):
        dx = abs(x2 - x1)
        dy = abs(y2 - y1)
        sx = 1 if x1 < x2 else -1
        sy = 1 if y1 < y2 else -1
        err = dx - dy

        while True:
            self.set_pixel(y1, x1, char)
            if x1 == x2 and y1 == y2:
                break
            e2 = 2 * err
            if e2 > -dy:
                err -= dy
                x1 += sx
            if e2 < dx:
                err += dx
                y1 += sy

            
    def display(self):
        print("\n".join(["".join(row) for row in self.data]))
```

```python
from abc import ABC, abstractmethod
import math

# Implementing a common base class, which subclasses will share functions from
class shape(ABC):
    def area(self):
        pass
        
    def perimeter(self):
        pass

    def coordinates(self):
        pass

    def inside_shape(self, x, y):
        pass

    def overlaps(self):
        pass

    # New function to paint the shape onto a canvas
    def paint(self):
        pass

# RECTANGLE
class rectangle(shape):
    def __init__(self, length, width, x, y):
        self.__length = length
        self.__width = width
        self.__x = x
        self.__y = y

    # Computes the area
    def area(self):
        return self.__width * self.__length

    # Computes the perimeter
    def perimeter(self):
        return 2 * (self.__width + self.__length)

    # Computes the coordinates of the shape, starting from the bottom left corner so we are adding length and width
    def coordinates(self):
        bottom_left = (self.__x, self.__y)
        bottom_right = (self.__x + self.__length, self.__y)
        top_right = (self.__x + self.__length, self.__y + self.__width)
        top_left = (self.__x, self.__y + self.__width)
        
        return [bottom_left, bottom_right, top_right, bottom_left]

    # Confirms whether or not a coordinate lies inside a shape or not
    def inside_points(self, x, y):
        # A point must be greater than or equal to (x,y) but less than or equal to (x+length, y+width)
        return (self.__x <= x <= self.__x + self.__length) and (self.__y <= y <= self.__y + self.__width)

    # Confirms whether or not two shapes overlap with one another
    def overlaps(self, other):
        # Checks if the rectangle and another rectangle overlap
        if isinstance(other, rectangle):
            return not (self.__x + self.__length < other.print_x() or
                        self.__x > other.print_x() + other.__length or
                        self.__y + self.__width < other.print_y() or
                        self.__y > other.print_y() + other.__width)
        # Checks if the rectangle and a circle overlap
        elif isinstance(other, circle):
            return any(other.inside_points(px, py) for px, py in self.coordinates()) or \
                   any(self.inside_points(px, py) for px, py in other.coordinates())
        # Checks other shapes using pervious methods
        elif isinstance(other, triangle):
            return any(other.inside_points(px, py) for px, py in self.coordinates()) or \
                   any(self.inside_points(px, py) for px, py in other.coordinates())
        else:
            return False

    # Prints the shape onto a canvas
    def paint(self, canvas):
        canvas.h_line(self.__x, self.__y, self.__length)
        canvas.h_line(self.__x + self.__width - 1, self.__y, self.__length)
        canvas.v_line(self.__x, self.__y, self.__width)
        canvas.v_line(self.__x, self.__y + self.__length - 1, self.__width)
            
    # Retrieves the values of the data members
    def print_length(self):
        return self.__length

    def print_width(self):
        return self.__width

    def print_x(self):
        return self.__x

    def print_y(self):
        return self.__y

# CIRCLE
class circle(shape):
    def __init__(self, radius, x, y):
        self.__radius = radius
        self.__x = x
        self.__y = y

    # Computes the area
    def area(self):
        return math.pi * (self.__radius ** 2)

    # Computes the perimeter
    def perimeter(self):
        return 2 * math.pi * self.__radius

    # Computes the coordinates of the shape, starting from the center
    def coordinates(self):
        # Using a loop to find 16 points on the circle
        points = []
        num_points = 16
        for i in range(num_points):
            angle = (2 * math.pi / num_points) * i
            # Creating the points by adding the radius, which puts us on the outline and the increments the coordinates by the angle between points
            x = self.__x + self.__radius * math.cos(angle)
            y = self.__y + self.__radius * math.sin(angle)
            points.append((x, y))
        return points
        
    # Confirms whether or not a coordinate lies inside a shape or not
    def inside_points(self, x, y):
        # Computes the distance from the given point from the center point
        distance = math.sqrt((x - self.__x) ** 2 + (y - self.__y) ** 2)
        # The given point must be between the center of the circle and the outside (radius) (less than the radius)
        return distance <= self.__radius

    # Confirms whether or not two shapes overlap with one another
    def overlaps(self, other):
        # Checks if the circle and another circle overlap
        if isinstance(self, circle):
            # Modifying the distance formula to calculate the distance between the centers of the circles
            distance = math.sqrt((self.__x - other.print_x()) ** 2 + (self.__y - other.print_y()) ** 2)
            # If the distance between the centers is less than or equal to the two radii then the circles overlap
            return distance <= self.__radius + other.__radius
        # Checks other shapes using pervious methods
        elif isinstance(other, rectangle):
            closest_x = max(other.print_x(), min(self.__x, other.print_x() + other.print_length()))
            closest_y = max(other.print_y(), min(self.__y, other.print_y() + other.print_width()))
            # Calculate distance from circle center to this closest point
            distance = math.sqrt((closest_x - self.__x) ** 2 + (closest_y - self.__y) ** 2)
            # If the the distance is less than or equal to the radius then they overlap
            return distance <= self.__radius
        elif isinstance(other, triangle):
            return other.overlaps(self)
        else:
            return False

    # Draws the circle onto a canvas
    def paint(self, canvas):
        # Use midpoint circle drawing algorithm
        x_center, y_center = self.__x, self.__y
        r = self.__radius
        x, y = r, 0
        p = 1 - r
        self._draw_circle_points(canvas, x_center, y_center, x, y)
        while x > y:
            y += 1
            if p <= 0:
                p = p + 2*y + 1
            else:
                x -= 1
                p = p + 2*y - 2*x + 1
            self._draw_circle_points(canvas, x_center, y_center, x, y)
            
   # Draw symmetric points for a circle
    def _draw_circle_points(self, canvas, xc, yc, x, y):
        points = [(xc + x, yc + y), (xc - x, yc + y),
                  (xc + x, yc - y), (xc - x, yc - y),
                  (xc + y, yc + x), (xc - y, yc + x),
                  (xc + y, yc - x), (xc - y, yc - x)
                 ]
        for px, py in points:
            if 0 <= px < canvas.width and 0 <= py < canvas.height:
                canvas.set_pixel(py, px)
                
    # Retrieves the value of the data members
    def print_radius(self):
        return self.__radius

    def print_x(self):
        return self.__x

    def print_y(self):
        return self.__y

# TRIANGLE
class triangle(shape):
    def __init__(self, base, height, x, y):
        self.__base = base
        self.__height = height
        self.__x = x
        self.__y = y

    # Compute the area
    def area(self):
        return 0.5 * self.__base * self.__height
        
    # Compute the perimeter by using the side_length function
    def perimeter(self):
        side_length = math.sqrt((self.__base / 2) ** 2 + self.__height ** 2)
        return self.__base + 2 * side_length

    # Computes the coordinates of the shape, assuming we start at the left most corner
    def coordinates(self):
        left_vertex = (self.__x, self.__y)
        right_vertex = (self.__x + self.__base, self.__y)
        top_vertex = ((self.__x + self.__base) / 2, self.__y + self.__height)

        return [left_vertex,
                right_vertex,
                top_vertex]

    # Confirms whether or not a coordinate lies inside a shape or not
    def inside_points(self, x, y):
        # Similiar to the coordinates function, however this separates them into their respective placement (x or y)
        # Bottom left x and y values
        x1, y1 = self.__x, self.__y
        # Bottom right x and y values
        x2, y2 = self.__x + self.__base, self.__y
        # Top x and y values
        x3, y3 = self.__x + self.__base / 2, self.__y + self.__height

        # Then we calculate the area of the entire triangle
        area_main = abs((x1*(y2-y3) + x2*(y3-y1) + x3*(y1-y2)) / 2.0)
        # And compute the area of the triangles where we replace a point (1, 2, and 3) with the point we are testing
        area1 = abs((x*(y2-y3) + x2*(y3-y) + x3*(y-y2)) / 2.0)
        area2 = abs((x1*(y-y3) + x*(y3-y1) + x3*(y1-y)) / 2.0)
        area3 = abs((x1*(y2-y) + x2*(y-y1) + x*(y1-y2)) / 2.0)

        # The total area will equal the sub areas if the point is inside the triangle
        if (area_main == area1 + area2 + area3):
            return True
        else:
            return False

    # Confirms whether or not two shapes overlap with one another
    def overlaps(self, other):
        # Checks if the triangle and another triangle overlap
        if isinstance(other, triangle):
            return any(other.inside_points(px, py) for px, py in self.coordinates()) or \
                   any(self.inside_points(px, py) for px, py in other.coordinates())
        # Checks if the triangle and a rectangle overlap
        elif isinstance(other, rectangle):
            return any(self.inside_points(px, py) for px, py in [(other.print_x(), other.print_y()),
                                                                 (other.print_x() + other.print_length(), other.print_y()),
                                                                 (other.print_x(), other.print_y() + other.print_width()),
                                                                 (other.print_x() + other.print_length(), other.print_y() + other.print_width())
                                                                ])
        # Checks if the triangle and any other shape overlap using previous method
        elif isinstance(other, circle):
            return any(other.inside_points(px, py) for px, py in self.coordinates()) or \
                   self.inside_points(other.print_x(), other.print_y())
        else:
            return False

    # Draws a triangle on a canvas
    def paint(self, canvas):
        # Draw triangle using lines (this is a basic approach)
        x1, y1 = self.__x, self.__y
        x2, y2 = self.__x + self.__base, self.__y
        x3, y3 = self.__x + self.__base // 2, self.__y + self.__height
        canvas.line(x1, y1, x2, y2)
        canvas.line(x2, y2, x3, y3)
        canvas.line(x3, y3, x1, y1)
        
    # Retrieve the values of the data members
    def print_base(self):
        return self.__base

    def print_height(self):
        return self.__height

    def print_x(self):
        return self.__x

    def print_y(self):
        return self.__y
```

```python
class compoundshape(shape):
    def __init__(self):
        self.shapes = []

    def add_shape(self, shape):
        self.shapes.append(shape)

    def area(self):
        return sum(shape.area() for shape in self.shapes)

    def perimeter(self):
        return sum(shape.perimeter() for shape in self.shapes)

    def paint(self, canvas):
        for shape in self.shapes:
            shape.paint(canvas)

    def overlaps(self, other):
        for shape in self.shapes:
            if shape.overlaps(other):
                return True
        return False
```
